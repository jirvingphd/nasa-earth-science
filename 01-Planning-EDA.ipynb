## 1) Global Mean Sea Level (GMSL)



**Dataset Title:** MEaSUREs Global Mean Sea Level (GMSL) from TOPEX/Poseidon, Jason-1, Jason-2, and Jason-3, Version 2

**Description:**
This dataset provides global mean sea level trends derived from the TOPEX/Poseidon, Jason-1, Jason-2, and Jason-3 satellite missions. The data spans from 1992 to present, offering monthly measurements of sea level anomalies. It includes corrections for instrumental and environmental effects, ensuring high accuracy for climate research.

**Key Features:**
- Time Span: 1992 to present
- Temporal Resolution: Monthly
- Variables: Sea level anomalies, uncertainties, corrections

**Access and Documentation:**
- [Dataset Access](https://podaac.jpl.nasa.gov/dataset/MERGED_TP_J1_OSTM_OST_GMSL_ASCII_V51)
- [User Guide and Documentation](https://podaac.jpl.nasa.gov/dataset/MERGED_TP_J1_OSTM_OST_GMSL_ASCII_V51)

This dataset is essential for studying long-term sea level changes and their impact on coastal regions.

````python
## PRocess the raw data for Pandas
import pandas as pd

# Load the GMSL data
file_path = './data/GMSL_TPJAOS_5.1.txt'

header_lines = []
with open(file_path) as f:
    
    for i,l in enumerate(f.readlines()):
        if l.startswith('HDR'):
            header_lines.append(l.strip("HDR"))
            
            
            # print(f"Line{i}: {l}")
        else:
            # i = i-1
            print(f"First Non-Header Row (line {i}): {l}")
            break 
                    
header = "\n".join(header_lines)
print(header)
````

````python
# Extract the column names
column_description = header.split("column description")[-1].strip()#.split("column units:")[0]
column_lines = column_description.split("\n\n")
column_names = []
for l in column_lines:
    lstrip = l.strip()
    if len(lstrip)>0:
        if lstrip[0].isdigit():
            column_names.append(' '.join(lstrip.split(' ')[1:]))
            
column_names
````

````python
df = pd.read_csv(file_path,sep='\s+',#delim_whitespace=True,
                 skiprows=i, header=None)

#Display the first few rows of the dataset
df.columns = column_names

df.head()
````

#### Explanation of the column names
1. **altimeter type 0=dual-frequency 999=single frequency (ie Poseidon-1)**:
   - **Explanation**: This indicates the type of altimeter used. A value of `0` means a dual-frequency altimeter, while `999` signifies a single-frequency altimeter like the Poseidon-1.
   - **Altimeter**: A device on satellites used to measure the height of the Earth's surface or sea level.

2. **merged file cycle #**:
   - **Explanation**: This is the cycle number of the merged data file, representing a specific time period during which data was collected.
   - **Cycle**: A regular interval of time in which the satellite completes its observations.

3. **year+fraction of year (mid-cycle)**:
   - **Explanation**: The year and part of the year (in decimal form) when the data was collected. For example, `1992.5` represents the middle of 1992.
   - **Fraction of Year**: Indicates the specific time within the year when the data was recorded.

4. **number of observations**:
   - **Explanation**: The total number of individual measurements taken by the satellite during the cycle.
   - **Observations**: Data points collected by the satellite sensors.

5. **number of weighted observations**:
   - **Explanation**: The number of observations, adjusted based on their reliability or importance.
   - **Weighted Observations**: Measurements given different levels of importance or accuracy.

6. **GMSL (Global Isostatic Adjustment (GIA) not applied) variation (mm) with respect to 20-year TOPEX/Jason collinear mean reference**:
   - **Explanation**: The change in global mean sea level (GMSL), measured in millimeters, without adjusting for global isostatic adjustment (GIA). GIA refers to the movement of the Earth's crust.
   - **GMSL**: The average sea level across the globe.

7. **standard deviation of GMSL (GIA not applied) variation estimate (mm)**:
   - **Explanation**: The standard deviation, in millimeters, of the GMSL measurements without GIA adjustment. This indicates the variability or uncertainty in the measurements.
   - **Standard Deviation**: A measure of how spread out the data points are from the average.

8. **smoothed (60-day Gaussian type filter) GMSL (GIA not applied) variation (mm)**:
   - **Explanation**: The GMSL data, without GIA adjustment, has been smoothed using a 60-day Gaussian filter to reduce noise and show trends more clearly.
   - **Smoothed Data**: Data that has been processed to remove short-term fluctuations and highlight longer-term trends.

9. **GMSL (Global Isostatic Adjustment (GIA) applied) variation (mm) with respect to 20-year TOPEX/Jason collinear mean reference**:
   - **Explanation**: The change in GMSL, measured in millimeters, with adjustments for GIA. This gives a more accurate representation of sea level changes.
   - **GIA Applied**: Adjustments made for the movement of the Earth's crust.

10. **standard deviation of GMSL (GIA applied) variation estimate (mm)**:
    - **Explanation**: The standard deviation, in millimeters, of the GMSL measurements with GIA adjustment. This indicates the variability or uncertainty in the adjusted measurements.
    - **Standard Deviation**: A measure of how spread out the data points are from the average.

11. **smoothed (60-day Gaussian type filter) GMSL (GIA applied) variation (mm)**:
    - **Explanation**: The GMSL data, with GIA adjustment, smoothed using a 60-day Gaussian filter to reduce noise and show trends more clearly.
    - **Smoothed Data**: Data that has been processed to remove short-term fluctuations and highlight longer-term trends.

12. **smoothed (60-day Gaussian type filter) GMSL (GIA applied) variation (mm); annual and semi-annual signal removed**:
    - **Explanation**: The GMSL data, with GIA adjustment, smoothed with a 60-day Gaussian filter and further processed to remove annual and semi-annual variations (like seasonal changes).
    - **Annual and Semi-Annual Signal Removed**: Removing seasonal effects to focus on longer-term changes.

13. **smoothed (60-day Gaussian type filter) GMSL (GIA not applied) variation (mm); annual and semi-annual signal removed**:
    - **Explanation**: The GMSL data, without GIA adjustment, smoothed with a 60-day Gaussian filter and processed to remove annual and semi-annual variations.
    - **Annual and Semi-Annual Signal Removed**: Removing seasonal effects to focus on longer-term changes.

### Python Code to Load Data from S3

```python
import boto3
import pandas as pd

# Initialize a session using Amazon S3
s3 = boto3.client('s3')

# Define bucket and object prefix
bucket_name = "podaac-ops-cumulus-protected"
object_key_prefix = "MERGED_TP_J1_OSTM_OST_GMSL_ASCII_V51/"

# List objects in the S3 bucket with the specified prefix
response = s3.list_objects_v2(Bucket=bucket_name, Prefix=object_key_prefix)

# Download each object in the prefix
for obj in response.get('Contents', []):
    key = obj['Key']
    local_filename = key.split('/')[-1]
    
    s3.download_file(bucket_name, key, local_filename)
    print(f'Downloaded {local_filename}')

# Example to read one of the downloaded files (assuming it's a CSV)
data = pd.read_csv('path_to_downloaded_file.csv')
print(data.head())
```

### Explanation
- **Initialize S3 client**: Connect to S3 using `boto3`.
- **Define Bucket and Prefix**: Set the bucket name and object key prefix as specified.
- **List Objects**: Retrieve a list of objects in the bucket with the specified prefix.
- **Download Files**: Loop through the list of objects and download each file locally.
- **Read Data**: Load one of the downloaded CSV files into a Pandas DataFrame for further analysis.
